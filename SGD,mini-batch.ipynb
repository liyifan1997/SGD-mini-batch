```python
import random
import matplotlib.pyplot as plt

#用y = Θ1*x1 + Θ2*x2来拟合下面的输入和输出
input_x = [[1, 4], [2, 5], [5, 1], [4, 2]]  #训练输入值
output_y = [19, 26, 19, 20]    #训练输出值
theta = [1, 1]              #θ参数初始化
loss = 10                   #loss先定义一个数，为了进入循环迭代
lr = 0.01                   #学习率
eps =0.0001                 #精度要求
max_iters = 10000           #最大迭代次数
error = 0                   #损失值
iter_count = 0              #当前迭代次数
loss_curve = []
iter_curve= []

while loss > eps and iter_count < max_iters:    #当模型精度未达到要求或者未达到最大迭代次数时，继续迭代
    loss = 0
    i = random.randint(0, 3)    # 随机抽取一组样本
    j = (i + 1) % 4             # 抽取另一组样本，j=i+1
    pred_y0 = theta[0] * input_x[i][0] + theta[1] * input_x[i][1]  # 预测值
    pred_y1 = theta[0] * input_x[j][0] + theta[1] * input_x[j][1]  # 预测值

    theta[0] = theta[0] - lr * (1 / 2) * ((pred_y0 - output_y[i]) * input_x[i][0] + (pred_y1 - output_y[j]) * input_x[j][0])# 权值更新
    theta[1] = theta[1] - lr * (1 / 2) * ((pred_y0 - output_y[i]) * input_x[i][1] + (pred_y1 - output_y[j]) * input_x[j][1])

    for i in range(4):
        pred_y = theta[0] * input_x[i][0] + theta[1] * input_x[i][1]        # 预测值
        error = (1/(2*2)) * (pred_y - output_y[i])**2                       #损失值
        loss = loss + error                                                 #总损失值

    loss_curve.append(loss)
    iter_curve.append(iter_count)
    iter_count += 1   #迭代计数加1
    print('iter_count:', iter_count, 'loss:', loss)   #输出每一次迭代的损失值
print('final theta:', theta)    #输出最终的网络系数
print('final loss:', loss)     #输出最终的损失值
print('final iter_count:', iter_count)    #最终的迭代次数
plt.plot(iter_curve, loss_curve, linewidth=3.0,  label = ' mini-batch loss value ')
plt.xlabel('iter_count')
plt.ylabel('loss value')
plt.legend(loc='upper right')
plt.show()   #输出损失值曲线
```
